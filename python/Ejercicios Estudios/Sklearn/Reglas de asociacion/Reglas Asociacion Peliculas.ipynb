### Importamos las librerias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
```

### Importamos el dataset de movie

```python
dataset_movies = pd.read_csv('movies.csv')
dataset_ratings = pd.read_csv('ratings.csv')
```

### Analizamos el data set para construir el dataset final para sacar las recomendaciones

```python
dataset_movies.head()
```

```python
dataset_ratings.head()
```

### Unimos los dataset por la columna en comun movieId

```python
dataset_unido = pd.merge(dataset_ratings,dataset_movies, on="movieId")
dataset_unido
```

### Sacamos todos los registros con un rating menor a 4 y quitamos las columnas timestamp y genres

```python
filtro_rating = dataset_unido['rating'] >= 4
dataset_filtrado = dataset_unido[filtro_rating]
dataset_filtrado=dataset_filtrado.drop('timestamp',axis=1)
dataset_filtrado=dataset_filtrado.drop('genres',axis=1)
dataset_filtrado
```

### Sacamos los valores unicos de los usuer id y de los titulos

```python
userId=dataset_filtrado['userId'].unique()
userId.sort()
movies_title=dataset_filtrado['title'].unique()
```

### Creamos el dataset de salida con los datos en cero, los index con nuestros usuarios y las columnas las peliculas

```python
dataset_final=pd.DataFrame(data='0', index=userId, columns=movies_title)
dataset_final
```

### Necesitamos crear dos listas, una con los titulos de las peliculas y la otra con los id de los usuarios

```python
list_titulos=list(movies_title)
list_user=list(userId)
```

### Recorro el dataset original usuario por usuario y coloco en el dataset final si el usuario vio la pelicula

```python
error = 0 #Mide las veces que no encontro la pelicula
for user in list_user: # Uso la lista de usuarios para hacer el for
    # Hago el filtro para seleccionar todas las peliculas de un solo usuario
    filtro_user = dataset_filtrado['userId'] == user 
    dataset_user = dataset_filtrado[filtro_user]
    # Cuento la cantidad de peliculas de ese usuario particular 
    num_pelis=len(dataset_user)
    # Recorro con otro for cada una de las peliculas de este usuario en particular 
    for row in range(num_pelis):
        # Saco el nombre de cada pelicula
        title_user = dataset_user.iloc[row,3]
        if list_titulos.count(title_user)>0:
            # Obtengo la posicion de la columna donde se encuentra la pelicula
            col=list_titulos.index(title_user)
            try:
                #Voy moviendo el puntero del user 
                ptr=list_user.index(user)
                #Cambio el valor en el dataset final correspondiente al usuario y a la pelicula
                dataset_final.iloc[ptr,col]=1
            except Exception as error:
                print(type(error),col,title_user,user)
        else:
            error = error+1
dataset_final
```

### Para usar la funcion apriory de mlxtend necesitamos que los valores sean boleanos

```python
for user in list_user: # Uso la lista de usuarios para hacer el for
    for movie in list_titulos:
        # Saco el nombre de cada pelicula
        col=list_titulos.index(movie)
        row=list_user.index(user)
        if  dataset_final.iloc[row,col]==1:
            dataset_final.iloc[row,col]= True
        else:
            dataset_final.iloc[row,col]= False
dataset_final
```

### El proceso demora mucho por lo que importamos el dataset listo por cuestiones de tiempo 

```python
dataset_true_false = pd.read_csv('dataset_true_false.csv', sep=';',index_col = 0)
dataset_true_false
```

### Entrenamos el modelo con la funcion apriori de la libreria mlxtend

```python
f=apriori(dataset_true_false, min_support=0.2, use_colnames=True, max_len=2)
```

### Vemos las reglas generadas

```python
rules=association_rules(f,metric='lift',min_threshold=2)
rules.head()
```
