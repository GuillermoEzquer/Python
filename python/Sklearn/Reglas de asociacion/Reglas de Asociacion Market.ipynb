### Importamos las librerias necesarias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

### Importamos el dataset

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
dataset.head()
```

### Vemos la forma de nuestro dataset para ver como lo recorremos para ordenar las transacciones

```python
dataset.shape
```

### Recorremos el dataset ordenando cada una de las transacciones

```python
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
#transactions
```

### Entrenamos el algoritmo apriori pasandole las transacciones y los parametros que definamos

```python
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
results = list(rules)
#results
```

### Definimos una funcion que nos ayude a visualizar los resultado de manera mas ordenada

```python
def inspect(results):
    lh          = [tuple(result[2][0][1]) for result in results] 
    rh 			= [tuple(result[2][0][0]) for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(rh, lh, supports, confidences, lifts))
```

### Usamos la funcion anteriormente definida para crear un dataframe para visualizar los resultados

```python
inspect_results=pd.DataFrame(inspect(results),columns=['Ancedente','Consecuente','Soporte','Confianza','Lift'])
inspect_results
```
