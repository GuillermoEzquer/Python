```python
import random 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
#from sklearn.datasets import make_blobs
%matplotlib inline
import pandas as pd
cust_df = pd.read_csv("Segmentación_De_Clientes.csv")
cust_df.head()
```

```python
df = cust_df.drop('Address', axis=1)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
X = df.values[:,1:]
X = np.nan_to_num(X)
Clus_dataSet = StandardScaler().fit_transform(X)
Clus_dataSet
```

```python
clusterNum = 3
k_means = KMeans(n_clusters = clusterNum)

k_means.fit(X)
labels = k_means.labels_
print(labels)
```

```python
df["Clus_km"] = labels
df.head(5)
```

```python
df.groupby('Clus_km').mean()
```

```python
area = np.pi * ( X[:, 1])**2  
plt.scatter(X[:, 0], X[:, 3], s=area, c=labels.astype(np.float), alpha=0.5)
plt.xlabel('Edad', fontsize=18)
plt.ylabel('Ingreso', fontsize=16)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D 
fig = plt.figure(1, figsize=(8, 6))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

plt.cla()
#Eje X
ax.set_xlabel('Nivel Educativo',fontsize=16)
#Eje Y
ax.set_ylabel('Edad',fontsize=16)
#Eje Z
ax.set_zlabel('Ingreso',fontsize=16)

ax.scatter(X[:, 1], X[:, 0], X[:, 3], c= labels.astype(np.float))
```
